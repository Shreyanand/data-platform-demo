```python
# best security practice to avoid hard-coding access credentials
with open('/opt/app-root/src/s3-auth.txt', 'r') as file:
    lines = [line.rstrip() for line in file.readlines()]

s3_endpoint_url = lines[0]
s3_access_key = lines[1]
s3_secret_key = lines[2]
s3_bucket = lines[3]
del lines

[s3_endpoint_url, s3_bucket]
```

```python
with open('/opt/app-root/src/trino-auth.txt', 'r') as file:
    lines = [line.rstrip() for line in file.readlines()]

trino_user = lines[0]
trino_passwd = lines[1]
trino_host = 'trino-secure-odh-trino.apps.odh-cl1.apps.os-climate.org'
trino_port = 443
del lines

trino_user
```

```python
import boto3
# Create an S3 client
s3 = boto3.client(
    service_name="s3",
    aws_access_key_id=s3_access_key,
    aws_secret_access_key=s3_secret_key,
    endpoint_url=s3_endpoint_url,
)
```

```python
import pandas as pd
obj = s3.get_object(
    Bucket=s3_bucket,
    Key='urgentem/UrgentemDataSampleDec2020.csv')
dfSample = (pd.read_csv(obj['Body'])).convert_dtypes()
dfSample.head()
```

```python
dfSample.info(verbose=True)
```

```python
obj = s3.get_object(
    Bucket=s3_bucket,
    Key='urgentem/UrgentemDataSampleEmissionsTargetsDec2020.csv')
dfEmissions = (pd.read_csv(obj['Body'])).convert_dtypes()
dfEmissions.head()
```

```python
dfEmissions.info(verbose=True)
```

```python
dfEmissions.columns.to_list()
```

```python
rename_map = dict(list(zip(dfEmissions.columns.to_list(),[
    'company_name',
    'isin',
    'target_type',
    'scope',
    'coverage_s1',
    'coverage_s2',
    'coverage_s3',
    'reduction_ambition',
    'base_year',
    'end_year',
    'start_year',
    'base_year_ghg_emissions_s1',
    'base_year_ghg_emissions_s1_s2',
    'base_year_ghg_emissions_s3',
    'achieved_reduction'
])))
dfEmissions.rename(columns=rename_map, inplace=True)
```

```python
dfEmissions.info(verbose=True)
```

```python
# parquet has multiple options for appending or updating data
# including adding new files, or appending, sharding directory trees, etc
dfEmissions.to_parquet('/tmp/emissions.parquet', index=False)
s3.upload_file(
    Bucket=s3_bucket,
    Key='urgentem/trino/itr_emissions_sample/emissions.parquet',
    Filename='/tmp/emissions.parquet'
)
```
